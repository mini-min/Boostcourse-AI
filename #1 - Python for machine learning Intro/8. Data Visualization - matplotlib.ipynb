```python
import numpy as np
import matplotlib.pyplot as plt
```

## Scatter 

```python
x = np.random.rand(30)
y = np.random.rand(30)
color = np.random.rand(30)
area = np.pi * (30 * np.random.rand(30)) ** 2

plt.scatter(x, y, s=area, c=color, marker='o', alpha=0.5)
plt.show()
```

## Bar chart 

```python
data = [5, 20, 40, 10]

x = np.arange(4)
plt.bar(x, data, width=0.5, color='r')

plt.show()
```

## histogram

```python
x = np.random.randn(1000)
plt.hist(x, bins=100)
plt.show()
```

## Pie Chart

```python
x = [34, 32, 16, 18]
labels = ['A', 'B', 'C', 'D']

plt.pie(x, labels=labels)
plt.show()
```
